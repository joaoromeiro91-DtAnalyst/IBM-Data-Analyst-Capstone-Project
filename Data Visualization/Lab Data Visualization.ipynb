**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

pd.set_option('display.max_columns', None)
# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
QUERY = """
    SELECT CompTotal
    FROM main
    WHERE CompTotal IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)

df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors="coerce")
df.dropna(subset=['CompTotal'], inplace=True)

plt.figure(figsize=(12, 6))
df.plot(kind='hist', bins=10, color='darkblue', edgecolor='grey')
plt.title('Total Compensation')
plt.xlabel('Compensation')
plt.ylabel('Count')
plt.show()
```

```python
## Removing the outliers from the data
Q1 = df['CompTotal'].quantile(0.25)
Q3 = df['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

bound_up = Q3 + 1.5*IQR
bound_low = Q1 - 1.5*IQR

filterDF = df[(df['CompTotal'] >= bound_low) & (df['CompTotal'] <= bound_up)]

plt.figure(figsize=(12, 6))
filterDF.plot(kind='hist', bins=40, color='darkblue', edgecolor='grey')
plt.title('Total Compensation')
plt.xlabel('Compensation')
plt.ylabel('Count')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
QUERY = """
    SELECT Age
    FROM main
    WHERE Age IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.value_counts()
```

```python
map_age = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}

df['Age_Num'] = df['Age'].map(map_age)
df = df.dropna(subset=['Age_Num'])

plt.figure(figsize=(12, 6))
plt.boxplot(df['Age_Num'], vert=False)
plt.title("Boxplot of Age of Respondents")
plt.xlabel("Age")
plt.grid(True)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = """
    SELECT Age, WorkExp
    FROM main
    WHERE Age IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x=df['Age'], y=df['WorkExp'], s=25, alpha=.75)
plt.title("ScatterPlot of Age vs Work Experience of Respondents")
plt.xlabel("Age")
plt.ylabel("Work Experience")
plt.xticks(rotation=45)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
QUERY = """
    SELECT TimeSearching, Frustration, Age
    FROM main
    WHERE TimeSearching IS NOT NULL AND Frustration IS NOT NULL AND Age IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
df['Frustration'] = df['Frustration'].astype(str).str.split(';')
clean_df = df.explode('Frustration')
clean_df['Frustration'] = clean_df["Frustration"].str.strip().str.replace(" ", "_")
```

```python
clean_df['Age'].value_counts()
```

```python
## here i take advantage of the variable map_age that is already declared in the boxplot part
clean_df['Age'] = clean_df['Age'].map(map_age)
clean_df['Age'].head()
```

```python
clean_df['TimeSearching'].value_counts()
```

```python
map_time = {
    "30-60 minutes a day": 45,
    "15-30 minutes a day": 22.5,
    "60-120 minutes a day ": 90,
    "Over 120 minutes a day": 135,
    "Less than 15 minutes a day":7.5
}

clean_df['TimeSearching'] = clean_df['TimeSearching'].map(map_time)
clean_df.head()
```

```python
scal_age = clean_df['Age'] / clean_df['Age'].max() *1000
plt.figure(figsize=(15, 6))
plt.scatter(
    y=clean_df['Frustration'],
    x=clean_df['TimeSearching'],
    s=scal_age,
    alpha=.45,
    c='skyblue',
    edgecolors="w", 
    linewidth=0.5
)
plt.ylabel('Frustation')
plt.xlabel('Time Searching')
plt.xticks(rotation=45)
plt.title('Frustation vs Time Searching for a solution with Age as Bubble size')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY= """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.value_counts()
```

```python
df['DatabaseWantToWorkWith'] = df['DatabaseWantToWorkWith'].str.split(";")
expl_df = df.explode('DatabaseWantToWorkWith')
expl_df['DatabaseWantToWorkWith'] = expl_df['DatabaseWantToWorkWith'].str.strip()
expl_df.value_counts()
```

```python
val_df = expl_df['DatabaseWantToWorkWith'].value_counts()
top5 = val_df[:5]
otherDB = val_df[5:].sum()
finalDb = pd.concat([top5, pd.Series({'Other':otherDB})])
plt.figure(figsize=(10, 10))
plt.pie(finalDb.values,
        labels=finalDb.index,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Preferred Databases to Work With (Top 5 + Others)")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
QUERY = """
    SELECT Age, TimeSearching, TimeAnswering
    FROM main
    WHERE Age IS NOT NULL
    AND TimeSearching IS NOT NULL
    AND TimeAnswering IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
df = df[df['Age'] == '25-34 years old']
```

```python
df['TimeSearching'].value_counts()
```

```python
df['TimeAnswering'].value_counts()
```

```python
time_map = {
    "Less than 15 minutes a day": 7.5,
    "15-30 minutes a day": 22.5,
    "30-60 minutes a day": 45,
    "60-120 minutes a day": 90,
    "Over 120 minutes a day": 135
}
df["TimeSearching_num"] = df["TimeSearching"].map(time_map)
df["TimeAnswering_num"] = df["TimeAnswering"].map(time_map)
```

```python
med = {
    "TimeSearching": df["TimeSearching_num"].median(),
    "TimeAnswering": df["TimeAnswering_num"].median()}

labels = ['30-35 Age Group']
searching = [med["TimeSearching"]]
answering = [med["TimeAnswering"]]

plt.bar(labels, searching, label='Time Searching')
plt.bar(labels, answering, bottom=searching, label='Time Answering')

plt.ylabel('Median Time (minutes)')
plt.title('Median Time Spent Searching vs Answering From Age 30–35')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY= """
    SELECT Age, CompTotal
    FROM main
    WHERE Age IS NOT NULL
    AND CompTotal IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
df['Age'].value_counts()
```

```python
age_group = ['45-54 years old', '55-64 years old']
df_filter = df[df['Age'].isin(age_group)]
df_filter['CompTotal'] = pd.to_numeric(df_filter['CompTotal'], errors='coerce')

median_comp = df_filter.groupby("Age")["CompTotal"].median().loc[age_group]


plt.figure(figsize=(8, 6))
median_comp.plot(kind="bar", color="skyblue", edgecolor="black", ylim=[80000,140000])

plt.title("Median Compensation by Age Group 45–64 years old")

plt.ylabel("Median Total Compensation")
plt.xlabel("Age Group")
plt.xticks(rotation=0.1)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = """
SELECT MainBranch
FROM main
WHERE MainBranch IS NOT NULL
""" 

df = pd.read_sql_query(QUERY, conn)
df
```

```python
df["MainBranch"].value_counts()
```

```python
mb_map = {'I am a developer by profession':'I am a developer \n by profession', 
          'I am learning to code':'I am learning \n to code',
          'I code primarily as a hobby':'I code primarily \n as a hobby',
          'I am not primarily a developer, but I write code sometimes as part of my work/studies':'I am not primarily \n a developer, \n but I write code  \n sometimes as part \n of my work/studies',
          'I used to be a developer by profession, but no longer am':'I used to be \n a developer by \n profession, \n but no longer am'
         }

df["MainBranch"] = df["MainBranch"].map(mb_map)
```

```python
counts = df["MainBranch"].value_counts()

plt.figure(figsize=(12, 6))


plt.barh(counts.index, counts.values, color='magenta', edgecolor="black")

plt.title("Type of answers for 'Why do you code?'")
plt.xlabel("Count")
plt.ylabel("Response")
plt.grid(True)
plt.tight_layout()
plt.show()
```
