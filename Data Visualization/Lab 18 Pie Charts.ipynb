### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
QUERY = """
    SELECT DatabaseWantToWorkWith
    FROM main
    WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df
```

```python
df['DatabaseWantToWorkWith'] = df['DatabaseWantToWorkWith'].str.split(';')
df_expl = df.explode('DatabaseWantToWorkWith')
df_expl['DatabaseWantToWorkWith'] = df_expl['DatabaseWantToWorkWith'].str.strip()
df_expl
```

```python
df_top5 = df_expl['DatabaseWantToWorkWith'].value_counts().head()
plt.figure(figsize=(12, 10))
plt.pie(x=df_top5.values, labels=df_top5.index, autopct='%1.1f%%')
plt.title('Top 5 Wanted Database to Work With')
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
QUERY = """
    SELECT DevType
    FROM main
    WHERE DevType IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df
```

```python
df['DevType'].unique()
```

```python
df_top5 = df['DevType'].value_counts().head()
plt.figure(figsize=(12, 8))
plt.pie(x=df_top5.values, explode=[.1, 0, 0, 0, 0], labels=df_top5.index, autopct='%1.1f%%', startangle=90,
       shadow=True)
plt.title('Top 5 of Developer Type')
plt.tight_layout()
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= 'main'
"""
df = pd.read_sql_query(QUERY, conn)
print(df.iat[0, 0])
```

```python
##Write your code here
QUERY = """
    SELECT "OpSysProfessional use"
    FROM main
    WHERE "OpSysProfessional use" IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df.rename(columns={"OpSysProfessional use": 'OpSysProf'}, inplace=True)
df
```

```python
df['OpSysProf'] = df['OpSysProf'].str.split(';')
df_expl = df.explode('OpSysProf')
df_expl['OpSysProf'] = df_expl['OpSysProf'].str.strip()
df_expl.value_counts()
```

```python
plot=df_expl['OpSysProf'].value_counts()
plt.pie(x=plot.values, explode=[0.1] * len(plot),  autopct='%1.1f%%', shadow=True,
        pctdistance=1.2, wedgeprops={'linewidth': 3}, textprops={'fontsize': 9})
plt.title('Operating System used Professionaly', loc='center')
plt.legend(plot.index, loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.axis('equal')
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here
QUERY = """
    SELECT LanguageAdmired
    FROM main
    WHERE LanguageAdmired IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)
df
```

```python
df['LanguageAdmired'] = df['LanguageAdmired'].str.split(';')
df_expl = df.explode('LanguageAdmired')
df_expl['LanguageAdmired'] = df_expl['LanguageAdmired'].str.strip()
df_expl
```

```python
plot = df_expl['LanguageAdmired'].value_counts().head(15)
size = len(df_expl['LanguageAdmired'].value_counts())
other = df_expl['LanguageAdmired'].value_counts().tail(size - 15)
other = other.values.sum()
other = pd.Series([other], index=['Other'])
plot = pd.concat([plot, other])
plot
```

```python
plt.figure(figsize=(12, 8))
plt.pie(x=plot.values, explode=[0.1] * len(plot),  autopct='%1.1f%%', shadow=True,
        pctdistance=1.1, wedgeprops={'linewidth': 3}, textprops={'fontsize': 12})
plt.title('IDE Tools most used', loc='center')
plt.legend(plot.index, loc='center left', bbox_to_anchor=(1, 0.5))
plt.axis('equal')
plt.show()
```
