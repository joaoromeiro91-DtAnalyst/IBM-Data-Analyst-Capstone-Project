### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
df_filter = df[['Age', 'SOPartFreq', 'JobSat']].dropna()
df_filter.head()
```

```python
df_filter['Age'].unique()
```

```python
df_filter['SOPartFreq'].unique()
```

```python
df_filter['JobSat'].unique()
```

```python
map_age = {
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '25-34 years old': 29.5,
    '55-64 years old': 59.5,
    '18-24 years old': 21,
    '65 years or older': 66,
    'Prefer not to say': None,
    'Under 18 years old': 17
}

df_filter['Age'] = df_filter['Age'].map(map_age)
```

```python
map_sofreq = {
    'Less than once per month or monthly': 'Less than once per\n month or monthly',
    'A few times per week': 'A few times per week',
    'A few times per month or weekly': 'A few times per \nmonth or weekly',
    'I have never participated in Q&A on Stack Overflow': 'I have never participated\n in Q&A on \nStack Overflow',
    'Daily or almost daily': 'Daily or almost daily',
    'Multiple times per day': 'Multiple times\n per day'
}
df_filter['SOPartFreq'] = df_filter['SOPartFreq'].map(map_sofreq)
```

```python
##Normalize the data for better view of the diferences of the sizes of the bubbles
df_filter['JobSat'] = (df_filter['JobSat'] - df_filter['JobSat'].min())/(df_filter['JobSat'].max() - df_filter['JobSat'].min())
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df_filter['SOPartFreq'], df_filter['Age'], s=(df_filter['JobSat']*1000))
plt.title('Times using Stack Overflow vs Age and Job Satisfation')
plt.xlabel('Frequency using Stack Overflow')
plt.ylabel('Age')
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
df_filter = df[['ConvertedCompYearly', 'JobSat', 'Age']].dropna()
df_filter.head()
```

```python
df_filter['ConvertedCompYearly'].value_counts()
```

```python
q1 = df_filter['ConvertedCompYearly'].quantile(.25)
q3 = df_filter['ConvertedCompYearly'].quantile(.75)
iqr = q3 - q1
bound_low = q1 - 1.5*iqr
bound_hig = q3 + 1.5*iqr
df_filter = df_filter[(df_filter['ConvertedCompYearly'] <= bound_hig) & (df_filter['ConvertedCompYearly'] >= bound_low)]
df_filter.head()
```

```python
df_filter['JobSat'].value_counts()
```

```python
df_filter['Age'].value_counts()
```

```python
map_age = {
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '25-34 years old': 29.5,
    '55-64 years old': 59.5,
    '18-24 years old': 21,
    '65 years or older': 66,
    'Prefer not to say': None,
    'Under 18 years old': 17
}

df_filter['Age'] = df_filter['Age'].map(map_age)
```

```python
size_age = df_filter['Age'].value_counts()
size_age = df_filter['Age'].map(size_age)
size_age = (size_age - size_age.min()) / (size_age.max() - size_age.min())
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df_filter['JobSat'], df_filter['ConvertedCompYearly'], s=size_age*100, alpha=.1)
plt.title('Relation between Job Satisfation and Compensation Yearly')
plt.xlabel('Job Satisfation')
plt.ylabel('Compensation Yearly')
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
df_filter = df[['LanguageHaveWorkedWith', 'Age']].dropna()
df_filter.head()
```

```python
df_filter['LanguageHaveWorkedWith'] = df_filter['LanguageHaveWorkedWith'].str.split(';')
df_expl = df_filter.explode('LanguageHaveWorkedWith')
df_expl['LanguageHaveWorkedWith'] = df_expl['LanguageHaveWorkedWith'].str.strip()
df_expl.head()
```

```python
size = df_expl['LanguageHaveWorkedWith'].value_counts()
size = df_expl['LanguageHaveWorkedWith'].map(size)
size = (size - size.min()) / (size.max() - size.min())
size
```

```python

plt.figure(figsize=(15, 9))
plt.scatter(df_expl['Age'], df_expl['LanguageHaveWorkedWith'],s=size*200, alpha=.01)
plt.title('Popularity of Programming Languages in the various Age Groups')
plt.xlabel('Age')
plt.ylabel('Programming Language')
plt.yticks(linespacing=1.5)
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
df_filter = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()
df_filter.head()
```

```python
df_filter['DatabaseWantToWorkWith'] = df_filter['DatabaseWantToWorkWith'].str.split(';')
df_expl = df_filter.explode('DatabaseWantToWorkWith')
df_expl['DatabaseWantToWorkWith'] = df_expl['DatabaseWantToWorkWith'].str.strip()
df_expl.head()
```

```python
size_db = df_expl['DatabaseWantToWorkWith'].value_counts()
size_db = df_expl['DatabaseWantToWorkWith'].map(size_db)
size_db = (size_db - size_db.min()) / (size_db.max() - size_db.min())
size_db
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(df_expl['JobSat'], df_expl['DatabaseWantToWorkWith'], s=size_db*200, c=size_db, alpha=.1)
plt.title('Relation Between Preferred Databases and Job Satisfation')
plt.xlabel('Preferred Databases')
plt.ylabel('Job Satisfation')
plt.yticks(linespacing=1.6)
plt.xticks(linespacing=1.6)
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
df_filter = df[['ConvertedCompYearly', 'DevType', 'JobSat']].dropna()
df_filter.head()
```

```python
##removing outliers
q1 = df_filter['ConvertedCompYearly'].quantile(.25)
q3 = df_filter['ConvertedCompYearly'].quantile(.75)
iqr = q3 - q1
bound_low = q1 - 1.5*iqr
bound_hig = q3 + 1.5*iqr
df_filter = df_filter[(df_filter['ConvertedCompYearly'] >= bound_low) & (df_filter['ConvertedCompYearly'] <= bound_hig)]
```

```python
df_filter['DevType'].unique()
```

```python
size_js = df_filter['JobSat'].value_counts()
size_js = df_filter['JobSat'].map(size_js)
size_js = (size_js - size_js.min()) / (size_js.max() - size_js.min())
size_js
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(df_filter['DevType'], df_filter['ConvertedCompYearly'], s=size_js*200, c=size_js)
plt.title('Relation between Developer type and Compensation and Job satisfation')
plt.xlabel('Developer Type')
plt.ylabel('Compensation')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
