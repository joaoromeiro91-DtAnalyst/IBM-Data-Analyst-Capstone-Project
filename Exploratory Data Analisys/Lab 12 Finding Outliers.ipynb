<h3>Step 1: Load and Explore the Dataset</h3>

<h3>Step 3: Identify High Compensation Outliers</h3>


Identify respondents with extremely high yearly compensation.

- Calculate basic statistics (mean, median, and standard deviation) for `ConvertedCompYearly`.

- Identify compensation values exceeding a defined threshold (e.g., 3 standard deviations above the mean).


```python
##Write your code here
df['ConvertedCompYearly'].describe()
```

```python
df["ConvertedCompYearly"].value_counts()
```

```python
df["ConvertedCompYearly"].isna().sum()
```

```python
df_comp = df['ConvertedCompYearly'].dropna()

mean_val = df_comp.mean()
median_val = df_comp.median()
std_val = df_comp.std()

threshold = mean_val + 3 * std_val

high_earners = df[df['ConvertedCompYearly'] > threshold]

print(f"The mean value is {mean_val}, the median value is {median_val} and the standard value is {std_val}.\n")
print(f"The Number of high earners are {len(high_earners)}. \n")
print(high_earners[['ConvertedCompYearly']].sort_values(by='ConvertedCompYearly', ascending=False).head(10))
```

<h3>Step 4: Detect Outliers in Compensation</h3>


Identify outliers in the `ConvertedCompYearly` column using the IQR method.

- Calculate the Interquartile Range (IQR).

- Determine the upper and lower bounds for outliers.

- Count and visualize outliers using a box plot.


```python
##Write your code here
df_comp.describe()
```

```python
Q1 = df_comp.quantile(0.25)
Q3 = df_comp.quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print(f"The IQR is: {IQR:,.2f}")
print(f"The lower bound is: {lower_bound:,.2f}")
print(f"The upper bound is: {upper_bound:,.2f}")

outliers = df_comp[(df_comp < lower_bound) | (df_comp > upper_bound)]
print(f"The number of outliers is: {len(outliers)}")
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(df_comp, orient='h')
plt.title("Boxplot of Yearly Compensation")
plt.xlabel("Compensation")
plt.show()
```

<h3>Step 5: Remove Outliers and Create a New DataFrame</h3>


Remove outliers from the dataset.

- Create a new DataFrame excluding rows with outliers in `ConvertedCompYearly`.
- Validate the size of the new DataFrame.


```python
##Write your code here
df_wo_outlin = df[(df['ConvertedCompYearly']>=lower_bound) & (df['ConvertedCompYearly']<=upper_bound)]
print(f"The old DataFrame as {df.shape[0]} rows")
print(f"The new DataFrame as {df_wo_outlin.shape[0]} rows")
```

<h3>Step 6: Correlation Analysis</h3>


Analyze the correlation between `Age` (transformed) and other numerical columns.

- Map the `Age` column to approximate numeric values.

- Compute correlations between `Age` and other numeric variables.

- Visualize the correlation matrix.


```python
##Write your code here
df['Age'].unique()
```

```python
age_map = {
    'Under 18 years old':17,
    '18-24 years old':21,
    '25-34 years old':29.5,
    '35-44 years old':39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}

df['Age_numeric'] = df['Age'].map(age_map)
```

```python
num_col = df.select_dtypes(include='number').columns
num_col
```

```python
cols = ['Age_numeric', 'CompTotal', 'WorkExp', 'ConvertedCompYearly', 'JobSat']
df_num = df[cols].dropna()
df_corr = df_num.corr()
print(df_corr)

plt.figure(figsize=(12, 6))
sns.heatmap(df_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
